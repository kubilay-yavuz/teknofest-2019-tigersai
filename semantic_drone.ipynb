```
import pickle

def load(path = './photos/semantic_drone_dataset/training_set/gt/bounding_box/bounding_boxes/person/imgIdToBBoxArray.p'):
    with open(path, 'rb') as pickleFile:
        return pickle.load(pickleFile)
bounding_bs=load()
```

```
"photos/semantic_drone_dataset/training_set/images"
df_csv=[]
for key in bounding_bs.keys():
    path="photos/semantic_drone_dataset/training_set/images/{}.jpg".format(key)
    for bbs in bounding_bs[key]:
        df_csv.append([path,int(bbs[0][0]),int(bbs[0][1]),int(bbs[1][0]),int(bbs[1][1]),"yaya"])
df_csv=pd.DataFrame(df_csv)
df_csv
```

```
df_csv.to_csv("semantic.csv",index=False,header=None)
```
