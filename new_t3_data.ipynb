```python
json_files=glob.glob("photos/t3_new_data/*_*_*_*")
df=[]
for json_file in json_files:
    with open(json_file,"r") as file:
        text=json.load(file)
    for image in text['data']:
        img_path='photos/t3_new_data/'+image['frame_link']
        for obj in image['objeler']:
            df.append([img_path,int(obj['x1']),int(obj['y1']),int(obj['x2']),int(obj['y2']),obj['tur']])
    print("Done: ",json_file)
df=pd.DataFrame(df)
df.to_csv("t3_new_data.csv",header=None,index=False)
```
