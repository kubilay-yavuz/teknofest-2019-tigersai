```python
new_df=new_df.reset_index(drop=True)
idx=list(new_df.index)
np.random.shuffle(idx)
new_df=new_df.iloc[idx]
new_df=new_df.reset_index(drop=True)
```

```python
a=new_df.loc[new_df[2]>=new_df[4]].index
b=new_df.loc[new_df[1]>=new_df[3]].index
a=np.append(np.array(a),np.array(b))
new_df=new_df.drop(a)
```
