```python
# send back response
# submit jsonu global yapıp frame_idleri dagıtıp submitle
def submit_jsons(text):
    code_200=[]
    code_400=[]
    code_500=[]
    for frame in text:
        r=session.post(url=url+"/api/cevap_gonder/", 
                json = frame)
        if r.status_code==200:
            code_200.append(frame["frame_id"])
        elif r.status_code==400:
            code_400.append(frame["frame_id"])
        elif r.status_code==500:
            code_500.append(frame["frame_id"])
        print("{}. frame resulted with: {}".format(frame["frame_id"],r.status_code))
```

```python
from multiprocessing import Process
procs = []
for index, frame in enumerate(all_json_subs_s):
    proc = Process(target=submit_jsons, args=frame)
    procs.append(proc)
    proc.start()

for proc in procs:
    proc.join()
```
