```python
import numpy as np
import urllib.request
import cv2
import json
import pandas as pd
import os
```

```python
os.listdir()
```

```python
file_path = 'veriler.json'
images_path = ''
nigg=[]
with open(file_path) as json_file:
    data = json.load(json_file)
    for jsn in data['frameler']:
        url = images_path + jsn['frame_url']
        objs = jsn['objeler']
        for o in objs:
            try:
                if(o['tur'] == 'yaya'):
                    text="yaya"
                    color = (0,0,255)
                elif o['tur']=="arac":
                    color = (255,0,0)
                    text="arac"
            except:
                color=(0,255,0)
                text="unidentified"
            csv_ex=[]
            csv_ex.append(url)
            csv_ex.append(int(o['x0']))
            csv_ex.append(int(o['y0']))
            csv_ex.append(int(o['x1']))
            csv_ex.append(int(o['y1']))
            csv_ex.append(text)
            nigg.append(csv_ex)
nigg=pd.DataFrame(nigg)
nigg.head()
```

```python
nigg.drop(nigg[nigg[5]=="unidentified"].index,inplace=True)
nigg
```

```python
nigg.to_csv("t3_data.csv",index=False,header=None)
```
